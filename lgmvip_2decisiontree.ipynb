```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics 
from sklearn.datasets import load_iris
```

```python
load_data = load_iris()
data = pd.read_csv(r"C:\Users\Baishali Ghosh\Downloads\IRIS.csv")
```

```python
data = data.values
X=data[:, 0:4]
Y=data[:,4]
X[0:4]
```

# Splitting Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

# Building Decision Tree Model

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

clf.fit(X_train,Y_train)
```

```python
#Predict the response for test dataset
Y_pred = clf.predict(X_test)
```

# Evaluating Model

```python
# Model Accuracy, how often is the classifier correct?

print("Accuracy Of Decision Tree:",metrics.accuracy_score(Y_test, Y_pred))
```

# Visualizing Decision Trees(Optimized)

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus


dot_data= StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True, feature_names = load_data.feature_names,class_names=load_data.target_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('species.png')
Image(graph.create_png())
```
